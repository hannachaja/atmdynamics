### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields {u,v} vary in 2D space {x,y}.

* Consider all possible LINEAR dependences of **V** = {u,v} on space (x,y) around an origin (0,0)
    * Measured by 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (that is, not varying with x,y) vector field U,V.

In other words, our *displayed* vector field V(x,y) = {u,v}(x,y) is:

$$ u_{disp} = U + c_1 x  +c_2 y$$
$$ v_{disp} = V + c_3 x  +c_4 y$$

## Exercise 2: Create and display the basic fields in Python
* The first code cell (shift-return to execute it and move on) sets up a coordinate system (x,y) from -10 to 10.
* The second cell defines the basic set of "pure" div, vort, def1, def2, and mean flow (U and V) fields. **Note that each field has a u and a v component.**
* Study the names and understand their logic! You will need it, in order to construct sums of these below.
* Finally, execute the cell that displays the vector fields as arrows using the *quiver* function.
* The first one has been done for you, then you complete the work for the other cases by example.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
psi = x*y
plt.contour(psi,20); plt.colorbar()
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

# First combination: pure divergence, plus a little mean U wind

```python
# Define the display field udisp,vdisp and give it a label.

udisp,vdisp = 12*uU + uvor, 12*vU + vvor
label = 'Pure unit vorticity plus U=12'

# Display the vector field as arrows
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

#Pure vorticity

```python
# Define the display field udisp,vdisp and give it a label.
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

# Pure vorticity of unit magnitude
# uvor,vvor = -1*y, 1*x

udisp,vdisp = uvor, vvor
label = 'Pure unit vorticity'


# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

# Pure deformation type 1

```python
# Define the display field udisp,vdisp and give it a label.
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

# Pure def1 of unit magnitude
# udef1,vdef1 = 1*x, -1*y

udisp,vdisp = udef1, vdef1
label = 'Pure unit def1'


# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

# Pure deformation type 2


```python
# Define the display field udisp,vdisp and give it a label.
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

# Pure def2 of unit magnitude
#udef2,vdef2 = 1*y, 1*x

udisp,vdisp = udef2, vdef2
label = 'Pure unit def2'


# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

# Sum of def1 + def2 (equal parts, or 2 to 1, or 1 to 2). Experiment!


```python
# Define the display field udisp,vdisp and give it a label.
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

udisp,vdisp = 1*udef1 + 1*udef2, 1*vdef1 + 1*vdef2
label = '50% def1 + 50% def2'


# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

# Sum of vorticity plus deformation in equal measure

Note that these are two curved flow patterns. Together they makes a flow we could describe as... shearing deformation!

```python
udisp,vdisp = uvor+ (0.71*udef1+0.71*udef2), vvor+ (0.71*vdef1+0.71*vdef2)
udisp,vdisp = uvor+ udef1, vvor+ vdef1
label = 'vorticity plus def1'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

# Exercise 3:

What are the **units and typical magnitude in the atmosphere** of these horizontal kinematic quantities (divergence, vorticity, deformation)?

To answer this, suppose u,v are in m/s (-10 to 10 range of values), and x,y are in 100s of km (so the *quiver* plots are about 2000 km across). You can just estimate the order of magnitude of anoy one of the derivatives (or a coefficient $c_1, c_2, c_3, c_4$). Express your answer using seconds as the time unit, and then convert to days as the time unit.  (You may take $1 day \simeq 10^5 s$ for convenience)


# Answer 3:

Typical values of the top and bottom of one of these kinematic derivative components are:  
$\delta u \simeq 10 m/s = 10m/s * 1km/1000m = 0.01km/s$

$\delta x \simeq 1000 km$

Therefore, the typical value $\delta u / \delta x$ is 0.01km/s / 1000km = $1*10^{-5}$ $s^{-1}$ \\
or ${1*10^-5}s^{-1}$ * $10^5s$/day = 1 $day^{-1}$.
